#### Encoding

- Encoding means converting Categorical data to Numerical data
- Because the ML expects the data in numerical format
- ML model developed by Math algorithms
- It is very very important to apply Encoding techniques
- The following methods are:
  - map
  - np.where
  - one hot encoding
  - Lable encoder

# map

- map is one method to convert Categorical data to numerical data
- Take one categorical column. Get the unique labels first
- Make a dictionary with assign number to each label
- For ex: case_status has 2 labels (unique values): 1.certified and 2.denied
- Assign 0 to certified
        1 to denied
- Create a dictionary with Labels as keys and numbers as values
- Ex: {'Certified':0,'Denied':1}

```python
case_status_df = visa_df['case_status']  #Read the column 
unique_status = case_status_df.unique()   # Get the unique values
dict_status = {'Certified':0,'Denied':1} # Make a dictionary
visa_df['case_status_new'] = case_status_df.map(dict_status)# Assign the values
visa_df
```

```python
visa_df = pd.read_csv(f"Visadataset.csv")
case_status_df = visa_df['case_status']
unique_status = case_status_df.unique()
unique_status.sort()
dict_status = {}
for i in range(len(unique_status)):
    dict_status[unique_status[i]]=i
dict_status
visa_df['case_status_new'] = case_status_df.map(dict_status)# Assign the values
visa_df
```

```python
visa_df = pd.read_csv(f"Visadataset.csv")
full_time_position_df = visa_df['full_time_position']
labels = full_time_position_df.unique()
labels.sort()
dict1 = {}
for i in range(len(labels)):
    dict1[labels[i]] = i
visa_df['full_time_position_new'] = full_time_position_df.map(dict1)
```

```python
 # Encoding map method for Visa data set visa_df. Perform the Map method for all categorical columns
visa_df = pd.read_csv(f"Visadataset.csv")
cat_cols = visa_df.select_dtypes(include='object').columns
for c in cat_cols:
    df = visa_df[c]
    labels = df.unique()
    labels.sort()
    dict1 = {}
    for i in range(len(labels)):
        dict1[labels[i]] = i
    visa_df[f"{c}_new"] = df.map(dict1)
visa_df
```

```python
# Encoding map method for Visa data set visa_df. Perform the Map method for case_status column
# Create a dictionary
visa_df = pd.read_csv(f"Visadataset.csv")

unique_labels = visa_df['case_status'].unique()
list1 = [i for i in range(len(unique_labels))]
unique_labels.sort()

# dict1 = {}
# for key,value in unique_labels,list1:
#     dict1[key] = value
# dict1   

dict1 = dict(zip(unique_labels,list1))

visa_df['case_status'] = visa_df['case_status'].map(dict1)
visa_df
```

```python
# Encoding map method for Visa data set visa_df. Perform the Map method for all categorical columns
visa_df = pd.read_csv(f"Visadataset.csv")
cat_cols = visa_df.select_dtypes(include='object').columns
for c in cat_cols:
    df = visa_df[c]
    labels = df.unique()
    labels.sort()
    list1 = [i for i in range(len(labels))]
    dict1 = dict(zip(labels,list1))
    visa_df[c] = df.map(dict1)
visa_df
```

#### Label Encoder

- Label Encoder is a method to convert Categorical data to Numerical data
- It is same way as we did above
- We explored for loop - map method. Label Encoder also will do same
- It si under **scikit-learn** . We usually called it as **sklearn**
- under sklearn we have **class** called as **preprocessing**
- Because currently we are doing data preprocessing only
- Under preprocessing class, we hae a method called as **LabelEncoder**
- sklearn
  - preprocessing
     - LabelEncoder()

- Any sklearn package has 3 steps:
  - Step1: Read the package
  - Step2: Save the package
  - Step3: Apply the fit transform on data




```python
visa_df = pd.read_csv(f"Visadataset.csv")
```

```python
# Read the package
from sklearn.preprocessing import LabelEncoder

# Save the package / Call the method
le = LabelEncoder()

# Apply the fit transform on data
visa_df['case_status'] = le.fit_transform(visa_df['case_status'])
visa_df
```

```python
# Apply fit_transform for all categorical columns
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

visa_df = pd.read_csv(f"Visadataset.csv")
cat_cols = visa_df.select_dtypes(include='object').columns

for c in cat_cols:
    visa_df[c] = le.fit_transform(visa_df[c])
visa_df
```

#### fit and transform

- fit - means develop the logic
- transform - means process the logic
- If we observe in map(), first we developed dictionary logic to store Lables, values
- Then we applied that dictionary to column using map()
- Developing the dictionary is kind of **fit**
- Apply the dictioneryto column means **transform** the data Categorical to Numerical
- Whenever we are changing or transforming the data then we use **fit_transofrm()**
- If we are not changing the data, then only use **fit**
  

#### np.where()

- Using np.where() also we can change teh categorical data to Numerical data
- But np.where() is used only for binary conditions
- np.where is same as **if and else**
- if condition is True, then apply True value
- if condition is False, then apply False value
- So np.where is used for binary labels. Column should have only 2 unique labels

```python
# apply np.where for case_status column. 
visa_df = pd.read_csv(f"Visadataset.csv")
case_status_labels = visa_df['case_status'].unique()
visa_df['case_status'] = np.where(visa_df['case_status']=='Certified',0,1)
visa_df
```
