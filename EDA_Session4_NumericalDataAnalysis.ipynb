#### Empirical rule

- All observed data will fall within 3 standard deviations from the mean
- 68% of the data will fall within one standard deviation, 95% within two standard deviations, and 99.7% within three standard deviations from the mean
- One standard deviation (Âµ Â± 1Ïƒ) = 68%- 
Two standard deviations (Âµ Â± 2Ïƒ = 95%
- 
Three standard deviations (Âµ 3Ïƒ3 = 97%
  .6

```python
# Get the 1 standard deviation of prevailing_wage

print(f"(Âµ Â± 1Ïƒ) = 68% = {len(visa_df['prevailing_wage'])*68/100}")

# Calculate the mean of 'prevailing_wage'
mean_wage = np.mean(visa_df['prevailing_wage'])
#  Calculate the Std of 'prevailing_wage'
std_wage = np.std(visa_df['prevailing_wage'])
# Get the lower and upper bounds for 1 standard deviation ((Âµ Â± 1Ïƒ))
lb = mean_wage - (1*std_wage)
ub = mean_wage + (1*std_wage)

# Get the visa data which is above lb and lower ub
cond1 = visa_df['prevailing_wage']>lb
cond2 = visa_df['prevailing_wage']<ub
# Apply the 'and' condition, then get the data for 1 standard deviation ((Âµ Â± 1Ïƒ))
cond3 = cond1 & cond2
visa_df[cond3]
```

```python
# Get the 2 standard deviation of prevailing_wage

print(f"(Âµ Â± 2Ïƒ) = 95% = {len(visa_df['prevailing_wage'])*95/100}")

# Calculate the mean of 'prevailing_wage'
mean_wage = np.mean(visa_df['prevailing_wage'])
#  Calculate the Std of 'prevailing_wage'
std_wage = np.std(visa_df['prevailing_wage'])
# Get the lower and upper bounds for 1 standard deviation ((Âµ Â± 2Ïƒ))
lb = mean_wage - (2*std_wage)
ub = mean_wage + (2*std_wage)

# Get the visa data which is above lb and lower ub
cond1 = visa_df['prevailing_wage']>lb
cond2 = visa_df['prevailing_wage']<ub
# Apply the 'and' condition, then get the data for 1 standard deviation ((Âµ Â± 2Ïƒ))
cond3 = cond1 & cond2
visa_df[cond3]
```

```python
# Get the 3 standard deviation of prevailing_wage

print(f"(Âµ Â± 3Ïƒ) = 97% = {len(visa_df['prevailing_wage'])*97/100}")

# Calculate the mean of 'prevailing_wage'
mean_wage = np.mean(visa_df['prevailing_wage'])
#  Calculate the Std of 'prevailing_wage'
std_wage = np.std(visa_df['prevailing_wage'])
# Get the lower and upper bounds for 1 standard deviation ((Âµ Â± 3Ïƒ))
lb = mean_wage - (3*std_wage)
ub = mean_wage + (3*std_wage)

# Get the visa data which is above lb and lower ub
cond1 = visa_df['prevailing_wage']>lb
cond2 = visa_df['prevailing_wage']<ub
# Apply the 'and' condition, then get the data for 1 standard deviation ((Âµ Â± 3Ïƒ))
cond3 = cond1 & cond2
visa_df[cond3]
```

#### Histogram

- By applying Empirical rule, we observed slight deviation in the count of data
- Also Mean is slightly deviated from Median
- This gives an indication that data is not fall under normal distribution
- To understand more clearly, will visualize the data
- It is numerical data. So the plot name is **Histogram**
- **Histogram** is a graph between Class Intervals and Class Interval Frequency
- The plot is available in matplot lib package
- method = **hist()**
  

```python
# Read the wage
# plt.hist()
wage = visa_df['prevailing_wage']
plt.hist(wage,bins=20)
```

```python
- Histogram will return 3 values
- Interval, count, array
- we have 20 intervals , so we have 20 interval counts
- bins
- bins means number of intervals
- we have 20 intervals so we need 21 values
- will display how many intervals are avialable
- For example in the bins array the first two forms one interval
- 2.136 to 15962.543 the count is 3863
- 3863 observation are available between 2.13670000e+00, 1.59625434e+04
- silmilarly 2175 observations are avialble between (1.59625434e+04, 3.19229500e+04)

```
