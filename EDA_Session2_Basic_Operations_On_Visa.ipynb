### Task - Using dtypes

- Create 2 lists
- One is Categorical columns list
- second is Numerical columns list

Whenever you want to create lists from Series, convert Series into a Dictionary
Iterate through dictionary items
  If item value is 'obect' then add respective item key to Categorical list
  Else add respective item key to Numerical list

```python
# Separate the columns which are categorical (obect) data type and numerical data type
dict_visa = dict(visa_df.dtypes)
catg_list = []
num_list = []
for key,value in dict_visa.items():
    if value=='object':
        catg_list.append(key)
    else:
        num_list.append(key)
print(catg_list)
print(num_list)
```

```python
# Separate the columns which are categorical (obect) data type and numerical data type
#Using List comprehension
dict_visa = dict(visa_df.dtypes)
cat_list = [i for i,j in dict_visa.items() if j=='object'] 
num_list = [i for i,j in dict_visa.items() if j!='obect']
print(cat_list)
print(num_list)
```

- Takes dtypes

```python
import pandas as pd

dict_visa = dict(visa_df.dtypes)
cat_list = [i for i,j in dict_visa.items() if j=='object']
num_list = [i for i,j in dict_visa.items() if j!='object']
print(len(cat_list))  #9
print(len(num_list))  #3
# We can't create a data frame, because 2 lists are not equal length
# If you want to create a DataFrame, then we need to make 2 lists to equal length
# We need to decide if making equal length is required or not. 
# In this situation, we can't make them equal length. So create 2 different DataFrames for 2 lists

df_cat = pd.DataFrame(cat_list,columns=['Categorical Column'])
df_num = pd.DataFrame(num_list,columns=['Numerical Column'])

# Save above DataFrames into CSV files
df_cat.to_csv('CategoricalColumns.CSV',index = False)
df_num.to_csv('NumericalColumns.CSV',index = False)
```

#### take()

- Syntax: .take(indices, axis: 'Axis' = 0, **kwargs) -> 'Self'
- Return the elements in the given *positional* indices along an axis.
- indices = An array of ints indicating which positions to take
- axis  = {0 or 'index', 1 or 'columns', None}, default 0

```python
visa_df.take([4]) 
# .take(indices, axis: 'Axis' = 0, **kwargs) -> 'Self'
# Return the elements in the given *positional* indices along an axis.
```

```python
visa_df.take([0,4],axis=1) 
```

```python
visa_df.take([1,4],axis=0) 
```

#### loc() and iloc()

- .iloc([< rows >,< columns >])
- .iloc([start:stop,  start:stop])
- .iloc([start:stop:step,  start:stop:step])

- .loc[row index]
- .loc[< row indeces list >,< Column names list>]
- .loc[<start:stop,< Start column name: End column name >]
- .iloc([start:stop:step,  < Column names >])

- .iloc[[2,5,7]]   - 2,5,7 indexed rows
- .iloc[:,[2,5,7]]  - All the rows and 2,5,7 indexed columns
- .iloc[:,2]        - All the rows, 2nd column. Returns Series type
- .iloc[:,[2]]      - All the rows, 2nd column. Returns DataFrame type

- Difference between loc and iloc:
- loc - We can provide the 'Column name' without counting index of the column
- iloc -We can't provide the 'Column name'. Should provide the column index to get the data
