```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import os
```

```python
# Read the Bank data
bank_df = pd.read_csv(f"bank.csv",sep=";")
bank_df
```

```python
# Convert Cat cols list
cat_cols = bank_df.select_dtypes(include='object').columns
cat_cols
```

```python
# Create a Directory Value_Counts_df in your python location using os.mkdirs
os.getcwd()
os.mkdir("Value_Counts_df")
```

```python
# Create Frequency table Data Frames in your PC
for i in cat_cols[1:]:
    cat_col_dict = bank_df[i].value_counts()
    cols = ['Labels','Data']
    cat_cols_df = pd.DataFrame(zip(cat_col_dict.keys(),cat_col_dict.values),columns = cols)
    cat_cols_df.to_csv("Value_Counts_df\\"+i+".csv",index=False)
```

```python
# Create a Directory Graphs_matplotlib
os.mkdir("Graphs_matplotlib")
```

```python
#  Read the DataFrames and plot the matplotlib graphs and save in directory
bank_cat_cols_df_list = os.listdir("Value_Counts_df")
for i in bank_cat_cols_df_list:
    df = pd.read_csv("Value_Counts_df\\"+i)
    plt.bar(cols[0],
           cols[1],
           data = df)
    xlabel = i.replace(".csv","")
    plt.xlabel(xlabel)
    plt.ylabel("Number of customers")
    plt.title(xlabel+" Barchart")
    plt.savefig("Graphs_matplotlib\\"+xlabel+".png")
    plt.show()
```

```python
# Create a Directory Bank_Graphs_Seaborn using os.mkdirs
os.mkdir("Bank_Graphs_Seaborn")
```

```python
# Plot the seaborn countplots and save in directory
cat_cols = bank_df.select_dtypes(include='object').columns
plt.figure(figsize=(15,5))
for i in cat_cols:
    order_cat_cols = bank_df[i].value_counts().keys()
    sbn.countplot(data = bank_df,
                  x = i,
                  order = order_cat_cols)
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.title(i+" Countplot")
    plt.savefig("Bank_Graphs_Seaborn\\"+i+".jpg")
    plt.show()
```
